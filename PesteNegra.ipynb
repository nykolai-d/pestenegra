#### Modelo de transmisión de la peste bubónica en la Edad Media

En el siglo XIV, tuvo lugar la famosa plaga de peste bubónica que mató a un tercio de la población de Europa. La enfermedad era causada por una bacteria trasmitida por la picadura de pulgas que habitaban mayormente en ratas. 

Recientemente, en un video viral sobre el coronavirus, el epidemiólogo Alfredo Miroli nombró como factor influyente en la expansión de la peste negra el movimiento religioso radical de los flagelantes. Estos eran fanáticos católicos que incurrían en prácticas de autoflagelación muy violentas para pedir perdón a Dios, pues veían la plaga como un castigo divino. Según Miroli, las carnes abiertas y la sangre atraían a las pulgas, convirtiéndose en focos de transmisión de la bacteria. Es decir, el hecho de formar parte del grupo de flagelantes aumenta el riesgo de contagio.

Consideremos una población de humanos $x$ y una población de pulgas $y$ infectada con la bacteria causante de la peste bubónica. Al igual que en el modelo de Lotka-Volterra, la población de humanos aumenta a menos de que entren en contacto con las pulgas y la población de pulgas disminuye a menos de que haya interacción con humanos. Además, supongamos que la probabilidad de contagio de la bacteria aumenta en función de la proporción de flagelantes en el grupo poblacional afectado:

$$\dfrac{dx}{dt} =x(a-by + f)$$

$$\dfrac{dy}{dt} =-y(c-dx)$$


Aquí $f$ es una constante positiva $0\le f \le 1$, que representa el porcentaje de individuos flagelantes en la población de humanos.

Con base en esta hipótesis y el modelo modificado, se presentan cuatro casos para análisis:

**CASO 0:** Modelo clásico de Lotka-Volterra

**CASO 1:** La tasa de crecimiento de las pulgas y los humanos es el mismo. El 10% de la población humana son flagelantes.

**CAS0 2:** La tasa de crecimiento de las pulgas es el doble respecto a la de los humanos. El 25% de la población humana son flagelantes.

**CAS0 3:** Las pulgas se reproducen 4 veces más rápido que los humanos. El 40% de la población humana son flagelantes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

"""
CASO 0: 
Modelo clásico de Lotka-Volterra
"""

a,b,c,d = 1, 1, 1, 1


def dX_dt(X,t):
    return [X[0]*(a-b*X[1]),-X[1]*(c-d*X[0])]
ts=np.linspace(0,12,100)
X0=[1.5,1.0]
Xs= odeint (dX_dt,X0,ts)
prey=Xs[:,0]
predators=Xs[:,1]
```

```python
plt.plot(ts,prey,'k',label="Humanos")
plt.plot(ts,predators,"r",label="Pulgas")
plt.xlabel("Tiempo")
plt.ylabel("Poblacion")
plt.title("CASO 0"); 
plt.legend();
```

En el modelo clásico de Lotka-Volterra, la gráficas de la tasa de crecimiento de la población humana y la de las pulgas tienen la misma amplitud y longitud de onda. Están desplazadas sobre el eje x, mostrando cómo cuando la población de pulgas está en su punto más alto, la población de humanos está descendiendo con una pendiente marcada. Asimismo, cuando la población de humanos está en su punto más alto, la población de pulgas está aumentando rápidamente. 

El  máximo y el mínimo de la población de pulgas es equivalente al máximo y al mínimo de la población de humanos, respectivamente. 

```python
plt.plot(prey,predators,"r")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 0");
```

```python
ic = np.linspace(1.0, 3.0, 21)
for r in ic:
    X0=[r,1.0]
    Xs=odeint(dX_dt,X0,ts)
    plt.plot(Xs[:,0],Xs[:,1],"-")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 0"); 
```

```python
"""
CASO 1: 
La tasa de crecimiento de las pulgas y los humanos es el mismo.
El 10% de la población humana son flagelantes.
"""

a,b,c,d,f = 1, 1, 1, 1, 0.1

def dX_dt(X,t):
    return [X[0]*(a-b*X[1]+f),-X[1]*(c-d*X[0])]
ts=np.linspace(0,12,100)
X0=[1.5,1.0]
Xs= odeint (dX_dt,X0,ts)
prey=Xs[:,0]
predators=Xs[:,1]
```

```python
plt.plot(ts,prey,'k',label="Humanos")
plt.plot(ts,predators,"r",label="Pulgas")
plt.xlabel("Tiempo")
plt.ylabel("Poblacion")
plt.title("CASO 1"); 
plt.legend();
```

En este modelo, las poblaciones de pulgas y humanos tienen la misma tasa de crecimiento. Si embargo, el $10\% $ de la población humana es flagelante. En consecuencia, el valor máximo de la gráfica de la población humana es siempre menor que el valor máximo de la población de pulgas. Sin embargo, el movimiento de las dos gráficas sigue pareciendo armónico a simple vista. Intuitivamente, podríamos aventurar la hipótesis de que la amplitud y la longitud de onda de ambas gráficas es el mismo (sería necesario probarlo). 

Este modelo es útil para ver el efecto que tienen los flagelantes en el comporamiento y la interacción de las dos poblaciones. Con tan solo el $10\%$ de la población flagelante, ya se observa una situación que parece desventajosa para la población humana.

```python
plt.plot(prey,predators,"r")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 1"); 
```

```python
ic = np.linspace(1.0, 3.0, 21)
for r in ic:
    P0=[r,1.0]
    Ps=odeint(dP_dt,P0,ts)
    plt.plot(Ps[:,0],Ps[:,1],"-")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 1"); 
```

```python
"""
CASO 2: 
La tasa de crecimiento de las pulgas es el doble de la tasa de crecimiento de los humanos.
El 25% de la población humana son flagelantes.
"""

a,b,c,d,f = 1, 1, 0.5, 1, 0.25


def dX_dt(X,t):
    return [X[0]*(a-b*X[1]+f),-X[1]*(c-d*X[0])]
ts=np.linspace(0,15,100)
X0=[1.5,1.0]
Xs= odeint (dX_dt,X0,ts)
prey=Xs[:,0]
predators=Xs[:,1]
```

```python
plt.plot(ts,prey,'k',label="Humanos")
plt.plot(ts,predators,"r",label="Pulgas")
plt.xlabel("Tiempo")
plt.ylabel("Poblacion")
plt.title("CASO 2"); 
plt.legend();
```

En este caso, la tasa de crecimiento de las pulgas es el doble respecto a la población humana; además, el $25\%$ de la población es flagelante. La tasa de crecimiento de las pulgas es pronunciado y, cuando alcanza su valor máximo, la población humana está acercándose a su valor mínimo. Esto significa que el aumento de la población de pulgas tiene un efecto drástico en la disminución de la población humana. 

Cuando la población humana está en su punto más bajo, la población de pulgas disminuye progresivamente. Cuando la población de pulgas alcanza un valor mínimo, la población humana repunta ligeramente. Pero tan pronto como esto ocurre, la tasa de crecimiento de las pulgas se dispara y hace que la población humana decrezca drásticamente una vez más. 

Es posible que, de los modelos aquí consignados, este sea el que más se acerque a representar el comportamiento real de la peste bubónica en la Edad Media. Sería interesante compararlo con datos reales.

```python
plt.plot(prey,predators,"r")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 2");
```

```python
ic = np.linspace(1.0, 3.0, 21)
for r in ic:
    P0=[r,1.0]
    Ps=odeint(dP_dt,P0,ts)
    plt.plot(Ps[:,0],Ps[:,1],"-")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 2"); 
```

```python
"""
CASO 3: 
Las pulgas se reproducen 4 veces más rápido que los humanos.
El 40% de la población humana son flagelantes.
"""

a,b,c,d,f = 1, 1, 0.25, 1, 0.4


def dX_dt(X,t):
    return [X[0]*(a-b*X[1]+f),-X[1]*(c-d*X[0])]
ts=np.linspace(0,30,100)
X0=[1.5,1.0]
Xs= odeint (dX_dt,X0,ts)
prey=Xs[:,0]
predators=Xs[:,1]
```

```python
plt.plot(ts,prey,'k',label="Humanos")
plt.plot(ts,predators,"r",label="Pulgas")
plt.xlabel("Tiempo")
plt.ylabel("Poblacion")
plt.title("CASO 3"); 
plt.legend();
```

Este modelo representa el caso extremo en el que las pulgas se reproducen 4 veces más rápido que los humanos y que el $40\%$ de la población humana es flagelante. La población de pulgas alcanza su punto máximo casi al mismo tiempo que la población humana alcanza su valor más bajo. Sin embargo, los humanos nunca se recuperan del todo del golpe, pues aún si la disminución de la población de pulgas les permite repuntar ligeramente, su número vuelve a decrecer drásticamente cuando la población de pulgas comienza apenas a remontar.

```python
plt.plot(prey,predators,"r")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 3");
```

```python
ic = np.linspace(1.0, 3.0, 21)
for r in ic:
    P0=[r,1.0]
    Ps=odeint(dP_dt,P0,ts)
    plt.plot(Ps[:,0],Ps[:,1],"-")
plt.xlabel("Humanos")
plt.ylabel("Pulgas")
plt.title("CASO 3"); 
```

$2.$ Para el modelo de Lotka-Volterra con crecimiento limitado

$$x^{\prime}=x(a-b y-\lambda x)$$
$$y^{\prime}=y(-c+d x-\mu y)$$

asigne valores a los parámetros y elabore el retrato de fase. ¿Qué diferencias observa respecto del modelo de Lotka-Volterra usual? ¿Cuál es la explicación biológica de dichas diferencias?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

a,b,c,d,g,m = 1, 1, 0.5, 1, 1, 1

def dX_dt(X,t):
    return [X[0]*(a-b*X[1]-g*X[0]),X[1]*(-c+d*X[0]-m*X[1])]
ts=np.linspace(0,30,100)
X0=[1.5,1.0]
Xs= odeint (dX_dt,X0,ts)
prey=Xs[:,0]
predators=Xs[:,1]
```

```python
plt.plot(ts,prey,'k',label="Humanos")
plt.plot(ts,predators,"r",label="Pulgas")
plt.xlabel("Tiempo")
plt.ylabel("Poblacion")
plt.title("Lotka-Volterra con crecimiento limitado"); 
plt.legend();
```

Hay diferencias en el movimiento de las gráficas: cuando la población de pulgas decrece, la de humanos también (esto no ocurría en el modelo clásico). Sin embargo, también se observa que ambas gráficas tienden a valores constantes después de un punto.

La explicación biológica a este fenómeno puede ser que las poblaciones de depredador-presa tienden a estabilizarse. Por ejemplo, en el caso de las pulgas y los humanos, hoy en día en una zona urbana como Bogotá hay un número más o menos estable de humanos y de pulgas. Ambas poblaciones sufren variaciones en su número (normales en el marco de sus ciclos vitales respectivos), pero no hay cambios drásticos como los observados en la Edad Media con la peste bubónica. 

Para que esto ocurriera, tendría que pasar algo salido de lo ordinario que rompiera con ese equilibrio poblacional relativo. Por ejemplo, el **coronavirus**.

El retrato de fase tiende a "abrirse", comenzando con figuras cerradas parecidas a esferas unas dentro de otras, convirtiéndose en polígonos irregulares con aristas entrecruzadas, hasta convertirse en un ángulo recto y estabilizarse ahí.
